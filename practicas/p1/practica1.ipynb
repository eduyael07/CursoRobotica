##### *2. ¿Qué es un archivo URDF?*
El formato URDF (Unified Robot Description Format) es una especificación XML que permite describir cómo es un robot; mediante elementos XML se pueden definir su aspecto visual, cinemática y dinámica. Robotica-facil-con-ros2
Cada parte rígida del robot (como una rueda o un brazo) se denomina link, y las conexiones entre dos links que permiten movimiento se llaman joints (articulaciones). Studocu
Una vez que se tiene el modelo URDF del robot, puede usarse para realizar simulaciones, por ejemplo con Gazebo o RViz.

##### *3. ¿Qué es git y github?*
**Git**: Git es un sistema de control de versiones distribuido, desarrollado por Linus Torvalds en 2005, que permite a los desarrolladores rastrear y gestionar cambios en el código de software. Su principal ventaja es la capacidad para gestionar múltiples ramas de desarrollo, lo que permite trabajar en nuevas características o correcciones de errores de manera aislada antes de fusionarlos en la rama principal. 
**GitHub**: GitHub es una plataforma basada en la web que utiliza Git como sistema de control de versiones. Permite a los desarrolladores colaborar en proyectos, gestionar problemas y realizar revisiones de código mediante "pull requests". K2WebHost
En resumen: Git permite gestionar el historial de proyectos localmente, mientras que GitHub es una plataforma en la nube que aloja repositorios de Git y facilita la colaboración en equipo.

##### *4. ¿Que es RVIZ?*
RViz es un visualizador 3D para el framework ROS (Robot Operating System). Permite visualizar el estado del robot y su entorno en tiempo real, incluyendo datos de sensores (como nubes de puntos, LiDAR, cámaras), estados del modelo del robot y otros tipos de información, facilitando la depuración y comprensión del comportamiento del sistema robótico.

### Ánalisis de resultados

¿Qué utilidad tiene el uso de un repositorio en línea para código?
> Un repositorio en línea permite almacenar, organizar y respaldar el código fuente de un proyecto de manera remota. Además, facilita el trabajo colaborativo entre varios desarrolladores, el control de versiones, el seguimiento de cambios y la recuperación de versiones anteriores del código. También ayuda a compartir proyectos y mantener una documentación centralizada.

¿Cuál es la convención de ángulos que utilizan los archivos URDF para los ángulos de las juntas respecto al eslabón padre?
> Los archivos URDF utilizan la convención de ángulos Roll-Pitch-Yaw (RPY) para definir la orientación de las juntas y eslabones respecto al eslabón padre. Los ángulos se expresan en radianes y siguen el orden de rotación sobre los ejes X (roll), Y (pitch) y Z (yaw).

¿Qué utilidad tiene describir un robot en un archivo URDF?
> Describir un robot en un archivo URDF permite representar su estructura física y cinemática de forma estandarizada. En el archivo se definen los eslabones, juntas, dimensiones, masas, sensores y propiedades visuales del robot. Esto facilita su simulación, visualización y control en herramientas como ROS y simuladores robóticos.
